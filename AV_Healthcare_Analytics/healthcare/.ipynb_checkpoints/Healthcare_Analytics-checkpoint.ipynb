```python
#convert string to 0,1,2.... numric values
from sklearn import preprocessing

encoder = preprocessing.LabelEncoder()

for i in dataset.columns:
    if isinstance(dataset[i][0], str):
            dataset[i] = encoder.fit_transform(dataset[i])
```

```python
dataset.head()
```

```python
dataset['Bed Grade'].value_counts()
```

```python
scaler=preprocessing.MinMaxScaler()
scaled=pd.DataFrame(scaler.fit_transform(dataset),columns=dataset.columns)
```

```python
x=dataset.drop(['case_id','Stay'],axis=1)
```

```python
y=dataset[['Stay']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.3, random_state = 8)
```

```python
from sklearn import preprocessing, utils
from sklearn import metrics 
```

# Decision Tree

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion = 'entropy') # default it will take gini
model.fit(x_train, y_train)
# print('Tranning score  :',round(model.score(x_train, y_train),4)*100,'%')
display(HTML(f'Training score  : <b>{round(model.score(x_train, y_train),5)*100}%</b>'))
#Predict Output
predicted= model.predict(x_test)
# print('Predicted score :',round(metrics.accuracy_score(y_test,predicted),2)*100,'%')
display(HTML(f'Predicted score : <b>{round(metrics.accuracy_score(y_test,predicted),3)*100}%</b>'))
print(predicted)
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted))
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,predicted)
cm 
```

```python
conf_matrix=pd.DataFrame(data=cm)
plt.figure(figsize = (15,5))
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu");
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```
